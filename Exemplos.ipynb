## **1.6 Método dropna()**
Um recurso que existe, e que, de certa forma, é bastante utilizado pelos programadores python é o método **dropna**.

O Método** dropna remove todas as linhas com valores NULL do DataFrame.** Ele age com eficiência, mas o problema é que ao remover as linhas, estes dados serão eliminados e perdidos.

```python
dbLIMPO = db.dropna()
display(dbLIMPO.head(8))
print("Nº de Registros = ", len(dbLIMPO))
```

Você pode aferir, p**ela exibição acima**, que ao exibir a quantidade de dados da base original com os indesejáveis NAN e ao exibir **a quantidade da nova base a gora tratada e sem nos indesejáveis NAN, o número de registros caiu consideravelmente;** isto é mais uma coisa indeseja e não apropriada de se fazer, pois não podemos nos dar ao luxo de perder dados importantes numa análise.
## **1.7 Método fillna()**
**Uma alternativa bastante eficaz para resolver o problema** dos indesejáveis NANs e ainda sim não perder os dados **é usar o método fillna**, **este método substitui todos os valores com NAN do dataframe por um valor específico colocado dentor do parênteses.** Ele é eficiente e melhor opção em relação ao dropna pois não ocasiona a perda de dados, mantendo todos os registros originais. Mas para utilizar o método filna é preciso fazer insto individualmente para cada coluna do seu objeto DataFrame.

```python
db['Linguagens_Conhecidas'].fillna('', inplace=True)
db['Grau_Conhecimento'].fillna('', inplace=True)
db['Expectativas'].fillna('', inplace=True)
db['Nao_gostaria'].fillna('', inplace=True)
display(db.head(10))
print("Nº de Registros = ", len(db))
```

```python
#import numpy as np
import matplotlib.pyplot as plt
#REF: https://www.geeksforgeeks.org/bar-plot-in-matplotlib/

RotulosEXP = ['Sim', 'Não']

def Grafico_Pizza(Rotulos, Quantias, Legenda, posExplode, LocLEG, Larg = 16, Alt = 9):
    # Rotulos: etiquetamento dos dados
    # Quantias: dados numéricos referente a cada rótulo
    # Legenda: etiquetamento da legenda
    # posExplode: posição na qual se encontra a fatia da pizza que se deseja ressaltar (explodir)
    # LocLEG: Localização onde será posicionada a Legenda do Gráfico (Ref: https://www.geeksforgeeks.org/change-the-legend-position-in-matplotlib/)

    #fig, ax = plt.subplots(figsize =(16, 9))
    fig, ax = plt.subplots(figsize =(Larg, Alt))
    explode = []
    for i in range(len(Rotulos)):
        if i !=posExplode:
            explode.append(0)
        else:
            explode.append(0.1)
    ax.pie(Quantias,
        explode=explode,
        labels=Legenda,
        autopct='%1.1f%%',
        shadow=True, startangle=90)
    ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    ax.legend(title='Disciplinas',
            loc=LocLEG,
            bbox_to_anchor=(1, 0, 0.5, 1))
    fig
```

```python
import numpy as np
import matplotlib.pyplot as plt

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]

# Set position of bar on X axis
br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, IT, color ='r', width = barWidth,
        edgecolor ='grey', label ='IT')
plt.bar(br2, ECE, color ='g', width = barWidth,
        edgecolor ='grey', label ='ECE')
plt.bar(br3, CSE, color ='b', width = barWidth,
        edgecolor ='grey', label ='CSE')
print("br1 = ", br1)
print("br2 = ", br2)
print("br3 = ", br3)

# Adding Xticks
plt.xlabel('Branch', fontweight ='bold', fontsize = 15)
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(IT))],
        ['2015', '2016', '2017', '2018', '2019'])

plt.legend()
plt.show()
```
